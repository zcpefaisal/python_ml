```python
df = pd.read_csv('data.csv')
```

```python
# average price of BMW cars with rounding 3
df[df.Make == 'BMW'].MSRP.mean().round(3)
```

```python
# subset of cars after year 2015
df[df.Year >= 2015]
```

```python
# get all the column missing value count
df[df.Year >= 2015].isnull().sum() 
```

```python
# get only Engine HP column missing value count
df[df.Year >= 2015].isnull().sum()['Engine HP']
```

```python
# get only Engine HP column missing value count
df[df.Year >= 2015]['Engine HP'].isnull().sum()
```

```python
# Calculate the average "Engine HP" in the dataset.
mean_hp = df['Engine HP'].mean().round(2)
mean_hp
```

```python
df['Engine HP'].fillna(0).mean()
```

```python
df['Engine HP'].fillna(mean_hp).mean()
```

```python
df[df.Make == "Rolls-Royce"]
```

```python
# Select only columns "Engine HP", "Engine Cylinders", "highway MPG".
df[df.Make=='Rolls-Royce'][["Engine HP","Engine Cylinders","highway MPG"]]
```

```python
# Select only columns "Engine HP", "Engine Cylinders", "highway MPG". and get unique value (remove duplicates)
df[df.Make=='Rolls-Royce'][["Engine HP","Engine Cylinders","highway MPG"]].drop_duplicates()
```

```python
# create new dataframe with then selected rolls-royce data
df_rr = df[df.Make=='Rolls-Royce'][["Engine HP","Engine Cylinders","highway MPG"]].drop_duplicates()
df_rr
```

```python
X = df_rr.values
X
```

```python
# Compute matrix-matrix multiplication between the transpose of X and X.
X.T.dot(X)
```

```python
# To get the transpose, use X.T
X.dot(X.T)
```
